```python

!pip install scikit-learn pandas numpy

# %%
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import pickle
import os


df = pd.read_csv('/kaggle/input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv')
print(df.head())

# %%
# Preprocessing
df['sentiment'] = df['sentiment'].map({'positive': 1, 'negative': 0})

# %%
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    df['review'], df['sentiment'], test_size=0.2, random_state=42
)

# %%
# Vectorization
vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)


# Model training
model = LogisticRegression(max_iter=5100)
model.fit(X_train_vec, y_train)


# Evaluation
print("Test accuracy:", model.score(X_test_vec, y_test))


# Save models
!mkdir -p /kaggle/working/models
with open('/kaggle/working/models/model.pkl', 'wb') as f:
    pickle.dump(model, f)
with open('/kaggle/working/models/vectorizer.pkl', 'wb') as f:
    pickle.dump(vectorizer, f)

# %%
# Download the models (run this in Kaggle)
from IPython.display import FileLink
FileLink('/kaggle/working/models/model.pkl')
FileLink('/kaggle/working/models/vectorizer.pkl')
```
